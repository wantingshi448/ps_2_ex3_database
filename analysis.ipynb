```python
# =============================================================================
# Cell 1 - Setup and Import (Enhanced Version)
# =============================================================================

"""
NBA Analysis Using Modular Functions
=====================================
This notebook demonstrates the use of reusable functions from nba/analysis.py
"""

# Import libraries
import sys
import os
import sqlite3

# Debug: Print current directory
print("ğŸ“‚ Current working directory:", os.getcwd())
print("ğŸ“‹ Files in current directory:", os.listdir('.'))

# Add project root to Python path
project_root = os.getcwd()
if project_root not in sys.path:
    sys.path.insert(0, project_root)
    print(f"âœ… Added to Python path: {project_root}")

# Verify nba folder exists
if 'nba' in os.listdir('.'):
    print("âœ… Found 'nba' folder!")
else:
    print("âŒ ERROR: 'nba' folder not found in current directory!")

# Import modular functions
try:
    from nba.analysis import (
        run_query,
        create_bar_chart,
        get_top_n_teams_by_home_wins,
        get_highest_scoring_games,
        analyze_data_quality
    )
    print("âœ… Successfully imported modular functions!")
except ImportError as e:
    print(f"âŒ Import Error: {e}")

# Connect to database
try:
    con = sqlite3.connect('data/nba.sqlite')
    print("âœ… Database connected successfully!")
except Exception as e:
    print(f"âŒ Database connection error: {e}")
```

```python
# =============================================================================
# Cell 2 - Fact 1: Teams with Most Home Wins
# =============================================================================

print("=" * 60)
print("FACT 1: Teams with Most Home Wins")
print("=" * 60)

# Use the modular function to get data
result = get_top_n_teams_by_home_wins(con, n=10)

# Display the data
print("\nğŸ“Š Top 10 Teams:")
print(result)

# Visualize using modular function
create_bar_chart(
    data=result,
    x_col='team_name_home',
    y_col='home_wins',
    title='ğŸ€ Top 10 Teams by Home Wins (1946-2024)',
    xlabel='Team Name',
    ylabel='Number of Home Wins',
    color='lightcoral'
)

# Print analysis
print(f"\nğŸ† Champion: {result.iloc[0]['team_name_home']}")
print(f"   Home Wins: {result.iloc[0]['home_wins']}")
print(f"\nğŸ“Š Difference between 1st and 2nd place:")
print(f"   {result.iloc[0]['home_wins'] - result.iloc[1]['home_wins']} wins")
```

```python
# =============================================================================
# Cell 3 - Fact 2: Highest-Scoring Game
# =============================================================================

print("=" * 60)
print("FACT 2: Highest-Scoring Game")
print("=" * 60)

# Get highest-scoring game
result = get_highest_scoring_games(con, n=1)

# Analyze data quality
analyze_data_quality(result)

# Extract game details
game = result.iloc[0]

print(f"\nğŸ”¥ Highest-scoring game in NBA history:")
print(f"   ğŸ“… Date: {game['game_date']}")
print(f"   ğŸ€ Matchup: {game['team_name_home']} vs {game['team_name_away']}")
print(f"   ğŸ“Š Score: {game['pts_home']}-{game['pts_away']}")
print(f"   ğŸ¯ Combined Total: {game['total_points']} points!")
print(f"   ğŸ† Winner: {game['team_name_home'] if game['wl_home'] == 'W' else game['team_name_away']}")
```
